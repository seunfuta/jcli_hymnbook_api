```python
import re
idval = ""
def get_id(text):
    #print(text[0:50])
    idmatch = re.search(r"\d+",text)
    if idmatch:
        idval = idmatch.group()
        return int(idmatch.group())
    else:    
        return 0
```

```python

def get_intro(text):
    intromatch = re.search(r'^.*?\n(?=1\.)', text, re.MULTILINE)
    if intromatch:
        return intromatch.group()
    else:
        return "no intro"
    


```

```python
def get_title(text):
    titlematch = re.search(r'1\..*?$', text, re.MULTILINE)
    if titlematch:
        return titlematch.group()[2:]
    else:
        return "no title"
```

```python

def get_language(text):
    langmatch = re.search(r'\d+y|e\.$', text, re.MULTILINE)
    if langmatch:
        #print(langmatch.group())
        y_or_e = re.search(r'y|e', langmatch.group())
        if y_or_e.group() == "y":
            return "yoruba"
        if y_or_e.group() == "e":
            return "english"
    else:
        return "no language"
```

```python

def get_group(text):
    group = ""
    return str(group)

```

```python
def get_tunelink(text):
    tunelink = ""
    return str(tunelink)

```

```python
def get_verses(text):
    versesmatch = re.findall(r'\d\.[^\d]+', text, re.MULTILINE)
    if versesmatch:
        verses = []
        for verse in versesmatch:
            verses.append(verse.split("Chr")[0])
        return verses
    else:
        return "no verses"

```

```python
def get_chorus(text):
    chorusmatch = re.search(r'Chr.\s+[^\d]+', text, re.MULTILINE)
    if chorusmatch:
        #print("chorusmatch")
        #print(chorusmatch.group())
        chorus_list = []
        choruses = chorusmatch.group().split("\n")
        #print("choruses length", len(choruses))
        for chorus in choruses:
            if re.match(r'Chr:.*',chorus,re.IGNORECASE):
                chorus = chorus[4:]
                chorus_list.append(chorus.strip())
            else:
                chorus_list.append(chorus.strip())
        return chorus_list
    else:
        return "no chorus"

```

```python
def get_addedChorus(text):
    chorusmatch = re.findall(r'Chr.\s+.*', text, re.MULTILINE)
    if chorusmatch:
        #print("chorusmatch")
        #print(chorusmatch)
    
        chorus_list = []
        for chorus in chorusmatch:
            chorus_list.append(chorus[4:].strip())
        
        if chorus_list[0][0:1] != chorus_list[-1][0:1]:
            #print("added chorus exists!")
            second_match = re.search(rf'{chorusmatch[-1]}.*', text, re.MULTILINE|re.DOTALL).group()
            if second_match:
                #print("added chorus found!")
                second_list = second_match.split("\n")
                third_list = []
                for xchorus in second_list:
                    if re.match(r'Chr:.*',xchorus,re.IGNORECASE):
                        xchorus = xchorus[4:]
                        third_list.append(xchorus.strip())
                    else:
                        third_list.append(xchorus.strip())
                return third_list
        else:
            return "no added chorus"
    else:
        return "no chorus"
```
